```python
#Importing necessary packages
import pandas as pd
# To display the plots even when the code has not been ran
%matplotlib inline 
import matplotlib.pyplot as plt # To plot the graphs
import seaborn as sns #Used to create the correlation matrix and Scatter Plot
```

```python
#load data using Pandas
iris = pd.read_csv("Iris.csv")
iris.info() #Gives description of the data
print("-------------------------------------------------")
```

Part a: Number of Tuples

```python
''' Determine the number of tuples
 Using the shape[0] we can get the number of rows or 
 tuples in a dataset '''
num_tuples = iris.shape[0]

print("The dataset has " + str(num_tuples) + " tuples.")
print("-------------------------------------------------")
```

Part b: Number of Features

```python
'''# Determine number of features, (name list)
 Using the shape[1] we can get the number of columns or 
 features in a dataset '''
num_features = iris.shape[1]

# Get the column names
feature_names = iris.columns
names = []
print("The dataset has " + str(num_features) + " features.")
for feature in feature_names:
    names.append(feature)
print("The feature names are: ")
print(names)
print("-------------------------------------------------")
```

Part c: Feature Data Types

```python
""" To determine feature datatype we can take advantage
    of the info() function which returns the features and 
    their datatypes """
iris.info()
print("-------------------------------------------------")
```

Part d: Number of Species

```python
''' Determine the number of unique species using the unique()
function. It will return the unique species into a list.
Then get the length of the list for the number
'''
species = iris['Class'].unique()
num_species = len(species)

print("The dataset has " + str(num_species) + " unique species.")
print("-------------------------------------------------")
```

Part e: Mean, Median, Quantiles and Range (max-min) for each Petal and Sepal Measurements

```python
# Creates a DataFrame to store display the results in a table
results = pd.DataFrame(columns=['Mean', 'Median', 'Q1', 'Q2', 'Q3', 'Range'])

# Calculate Mean, Median, Quantiles, and Range for each Petal and Sepal Measurement
for column in ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]:
    mean = iris[column].mean()
    median = iris[column].median()
    q1, q2, q3 = iris[column].quantile([0.25, 0.5, 0.75])
    range_ = iris[column].max() - iris[column].min() #max value minus min value

    ''' Append the results for this column to the results DataFrame.
    The .loc attribute adds the values as a row in the order specified
    '''
    results.loc[column] = [mean, median, q1, q2, q3, range_]

# Display the results
print(results)
print("-----------------------------------------------------")
```

Part f: Scatter Plot of Sepal Length vs Sepal Width

```python
# Creates a scatter plot of Sepal Length vs Sepal Width
plt.scatter(iris['Sepal Length'], iris['Sepal Width'])

# Title for the Scatter Plot
plt.title('Scatter Plot: Sepal Length vs Sepal Width')

# Label for the axes
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')


# Display the Plot
plt.show()
```

Part g: Histograms

```python
# Histograms for Sepal and Petal Measurements
iris.hist(bins=50, figsize=(20,15))
plt.show()
```

Part h: Boxplots

```python
# Boxplots for Sepal and Petal Measurements
iris.boxplot(figsize=(20,15))
plt.show()
```

Part i: Correlation Matrix and Correlation Scatter Plot

```python
# Create Correlation Matrix:
corr_matrix = iris.corr()

# Create a Correlation Scatter Plot
sns.pairplot(iris, hue='Class')

#Print the correlation matrix
print(corr_matrix)
print("-----------------------------------------------------\n")

# Show the plot
print("Correlation Scatter Plot:")
plt.show()
```
